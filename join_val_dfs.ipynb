```python
import pandas as pd
```

```python
inmet_val = pd.read_csv('inmet_val.csv')
focus_val = pd.read_csv('focus_val.csv')
```

```python
inmet_val['lat'] = inmet_val['lat'].str.replace(',', '.').astype(float)
inmet_val['lon'] = inmet_val['lon'].str.replace(',', '.').astype(float)
```

```python
inmet_val = inmet_val.apply(lambda x: x.fillna(x.median()) if x.isna().any() else x, axis=0)
inmet_val.isna().sum()
```

```python
focus_val['date'] = focus_val['date'].str[:13]
focus_val.head()
```

```python
inmet_val['date'] = inmet_val['DATA (YYYY-MM-DD)'] + ' ' + inmet_val['Hora UTC'].str[:2]
inmet_val['city'] = inmet_val['city_station']
inmet_val = inmet_val.drop(labels=["DATA (YYYY-MM-DD)", "Hora UTC", "city_station"], axis=1)
inmet_val.head()
```

```python
df_val = pd.merge(inmet_val, focus_val, on=['date', 'city'], how='inner')
df_val.head()
```

```python
df_val = df_val.drop(labels=["precipitation"], axis=1)
df_val = df_val.rename(columns={"state": "ESTADO",
                                "lat": "LATITUDE",
                                "lon": "LONGITUDE",
                                "date": "DATA",
                                "city": "CIDADE",
                                "biome": "BIOMA",
                                "days_without_rain": "DIAS SEM CHUVA",
                                "fire_risk": "RISCO DE FOGO"})
df_val.head()
```

```python
df_val.isna().sum()
```

```python
df_val.to_csv('df_val.csv', index=False)
```
